**Escojan una imagen a color, deberán aplicar las siguientes operaciones a las imágenes:**


*Muestren solo el canal azul de la figura*

```python
img = cv2.imread('mountain.jpg')
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(RGB_img)
```

```python
chanel_blue = RGB_img.copy() 

chanel_blue[:,:,0] = 0
chanel_blue[:,:,1] = 0

plt.imshow(chanel_blue)
```

*Cambien la imagen a escala de grises*

```python
gray_scale = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#img = cv2.imread('mountain.jpg')
plt.imshow(gray_scale, cmap='gray')
```

*Guarden una nueva imagen que contenga solo los canales rojo y verde*

```python
chanel_red_green = RGB_img.copy()

chanel_red_green[:,:,1] = 0

plt.imshow(chanel_red_green)
```

```python
cv2.imwrite('red_green.jpg', chanel_red_green)
```

*Guarden una nueva imagen que muestre la imagen en escala de grises*

```python
cv2.imwrite('black_white.jpg', gray_scale)
```
